### Pre-Processing

```python
# from my previous assignment
def contract(text):
    return contractions.fix(text)

def regTokenize(text):
    tok=RegexpTokenizer('[A-Za-z0-9]*[.]?\w+')
    return tok.tokenize(text) 

def lowercase(text):
    return text.lower()

def lemma(words):
    for i in range(0,len(words)):
        words[i]=WordNetLemmatizer().lemmatize(words[i])
    return words

def stemming(words):
    porter_stemmer=PorterStemmer()
    for i in range(0,len(words)):
        words[i]=porter_stemmer.stem(words[i])
    return words

def tweet(words):
    tok=TweetTokenizer()
    return tok.tokenize(words)

def comma(text):
    text = "".join(c for c in text if c not in ('!','.',':',',','"','?','(',')'))
    return text
```

### reading filenames and inserting into 'metadata' from files

```python
f1=open('names_1.txt','r')
out1=f1.readlines()

f2=open('names_2.txt','r')
out2=f2.readlines()

fnames=[]
titles=[]
for i in out1:
    s=i.split('   ')
    fnames.append('stories/'+s[0])
    word=s[1].split(' ',1)[1]
    titles.append(word.rstrip())

for i in out2:
    s=i.split('   ')
    fnames.append('stories/SRE/'+s[0])
    word=s[1].split(' ',1)[1]
    titles.append(word.rstrip())
    
metadata=[]
for i in range(len(fnames)):
    temp=[]
    temp.append(fnames[i])
    temp.append(titles[i])
    metadata.append(temp)
```

```python
for i in metadata:
    print(i)
    break
```

### creating title_dictionary

```python
title_dictionary={}
for i in metadata:
    doc_id=i[0]
    title_data=i[1]
    title_data=lowercase(title_data)
    title_data=contract(title_data)
    title_data=tweet(title_data)
    title_data=lemma(title_data)
    title_data=stemming(title_data)
    
    title_dictionary[doc_id]=title_data
```

```python
for key,value in title_dictionary.items():
    print(key,'-',value)
    break
```

### creating document body_dictionary vocabulary

```python
document_ids=[] #whole ids

body_terms=[]
body_words=[]

body_dictionary={}
# row=0
for i in metadata:
    docId=i[0] #fname
    print(docId)
    
    document_ids.append(docId)
    
    f=open(docId,'r+',errors='ignore')
    data=f.read()

    data=lowercase(data)
    data=contract(data)
    data=comma(data)
#     data=tweet(data)
    data=regTokenize(data)
    data=lemma(data)
    data=stemming(data)

    i.append(len(data)) #appending len(words in corresponding document) to that metadata row.
    data=set(data)
    data=list(data)
    
    for j in data:
        if(j not in body_dictionary.keys()):
            body_dictionary[j]=[]
            docfreq=1
            body_dictionary[j].append(docfreq)
            docs=[docId]
            body_dictionary[j].append(docs)
        else:
            body_dictionary.get(j)[1].append(docId)
            body_dictionary.get(j)[0]=len(body_dictionary.get(j)[1])
print('============================================================')
```

### pre-processing for test

```python
def pre_processing_test(term):
    term=lowercase(term)
    term=contract(term)
    term=comma(term)
#     term=tweet(term)
    term=regTokenize(term)
    term=lemma(term)
    term=stemming(term)
    return term
```

### Some useful functions

```python
def getDocLength(docId):#getting len(document)=number of words in that document from metadata
    for i in metadata:
        if(i[0]==docId):
#             print(i[0])
            return i[2]

def getWordFreqInBodyDoc(docId,word):#getting word frequency in (whole)Body zone(have title+body) in document
    f=open(docId,'r+',errors='ignore')
    data=f.read()
    data=lowercase(data)
    data=contract(data)
    data=comma(data)
#     data=tweet(data)
    data=regTokenize(data)
    data=lemma(data)
    data=stemming(data)
    
    freq=data.count(word)
    return freq

def getWordFreqinTitle(docId, word):#getting word frequency in Title zone(Heading zone) of the corresponding document
    return title_dictionary.get(docId).count(word)

def getQPosts(term):#getting postings of term from dictionary
    return body_dictionary.get(term)[1]

```

### Input

```python
query=input('enter query:')
k=int(input('enter k:'))
print(query,' ', k)
q_terms=pre_processing_test(query)
print(q_terms)

q_posts=[]
# for i in q_terms:
#     posts=getQPosts(i)
#     q_posts.append(posts)
    
for i in q_terms:
    if(i in body_dictionary.keys()):
        posts=getQPosts(i)
    else:
        print(i,' not in dictionary.')
        posts=[]
    q_posts.append(posts)
```

### getting intersection postings

```python
# total_postings=set.union(*map(set,q_posts))
total_postings=set.intersection(*map(set,q_posts))
total_postings=list(total_postings)
print(len(total_postings))
```

```python
n=len(document_ids)
print(n)
```

```python
title_weight=0.6
body_weight=0.4
scores=[]
for i in total_postings:
    print(i,'-->')
    
    total=0
    for q in range(0,len(q_terms)):
        
        bodyFreq=getWordFreqInBodyDoc(i,q_terms[q])
        titleFreq=getWordFreqinTitle(i, q_terms[q])
        
        if(bodyFreq>0 and titleFreq>0):
            freq=abs(bodyFreq-titleFreq)
            weight=1
        elif(bodyFreq>0 and titleFreq==0):
            freq=bodyFreq
            weight=0.4
        else:
            freq=titleFreq
            weight=0.6
        
        freq=bodyFreq
        tf=freq/getDocLength(i)
        
        inner=n/len(q_posts[q])
        idf=math.log(inner)
        
        tf_idf=tf*idf
        
        total=total+weight*tf_idf
#         print(q_terms[q],'-',bodyFreq,':', titleFreq,' Freq:',freq,' w:',weight,'-->',tf,'*',idf,'->',tf_idf,'weighted Tf-idf:',(tf_idf*weight))
    
    doc_tf=[i,total]
    scores.append(doc_tf)
```

```python
from operator import itemgetter
sorted_scores = sorted(scores, key=itemgetter(1),reverse=True)
```
