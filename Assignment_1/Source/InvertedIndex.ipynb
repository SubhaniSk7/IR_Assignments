```python
import nltk
import pandas as pd
import os

from collections import Counter
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize,RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk import wordpunct_tokenize
from nltk.stem import WordNetLemmatizer

import pandas as pd
import re as regex
import numpy as np
import scipy
import os,glob
from sklearn.externals import joblib
import contractions
```

```python
def contract(text):
    return contractions.fix(text)

def tokenize(text):
    words=word_tokenize(text)
    return words

def regTokenize(text):
    tok=RegexpTokenizer('[A-Za-z0-9]?\w+')
    return tok.tokenize(text)

def removePunct(text):
    words=wordpunct_tokenize(text)
    return words

def removePunctuations(words):
#     https://www.programiz.com/python-programming/examples/remove-punctuation
    punctuations='''!()-."[]".{};:'"\,<>./---::--...''``?@#$%^&*_~'''
    removed=[]
    for char in words:
        if char not in punctuations:
            removed.append(char)
        else:
            continue
    return removed
    
def lowercase(text):
    return text.lower()

def lemma(words):
    for i in range(0,len(words)):
        words[i]=WordNetLemmatizer().lemmatize(words[i])
    return words

def stemming(words):
    porter_stemmer=PorterStemmer()
    for i in range(0,len(words)):
        words[i]=porter_stemmer.stem(words[i])
    return words

def removestopwords(text):
    stopWords = set(stopwords.words('english'))
    wordsFiltered = []
    for w in text:
        if w not in stopWords:
            wordsFiltered.append(w)
    return wordsFiltered
```

```python
terms=[]
document_ids=[]

path = '20_newsgroups'
print(path)
for dire in glob.glob(os.path.join(path, '*')):
    print(dire)
    count=1
    for filename in glob.glob(os.path.join(dire, '*')):
        count+=1
        st=filename.split('\\')
#         print(st[2])
        s=st[1]+'/'
        s=s+st[2]

#         document_ids.append(st[2])
        document_ids.append(s)

        f=open(filename,'r+')
        data=f.read()

        data=lowercase(data)
        data=contract(data)
#         data=tokenize(data)
#         data=removePunctuations(data)
        data=regTokenize(data)
        data=lemma(data)
        data=stemming(data)
#         data=removestopwords(data)

        data=set(data)
        data=list(data)

        for i in range(0,len(data)):
            t=data[i]
#             id=st[2]
            id=s
            row=[]
            row.append(t)
            row.append(id)
            terms.append(row)
    print('============================================================')
```

```python
def ORing(x,y):
    print(x,' OR ',y)
    x_dict=di.get(x)
    y_dict=di.get(y)
    
    x_set=set(x_dict[1])
    y_set=set(y_dict[1])
    
    print('\nretrieved docs are:',x_set.union(y_set))
    print('\nno.of retrieved docs:',len(x_set.union(y_set)))

def ANDing(x,y):
    print(x,' AND ',y)
    x_dict=di.get(x)
    y_dict=di.get(y)
    
    x_set=set(x_dict[1])
    y_set=set(y_dict[1])
    
    print('\nretrieved docs are:',x_set.intersection(y_set))
    print('\nno.of retrieved docs:',len(x_set.intersection(y_set)))

def ANDNOT(x,y):
    print(x,' AND NOT ',y)
    dIds=set(docids)
    
    x_dict=di.get(x)
    y_dict=di.get(y)
    
    x_set=set(x_dict[1])
    y_set=set(y_dict[1])
    
    print('\nretrieved docs are:',x_set.difference(y_set))
    print('\nno.of retrieved docs:',len(x_set.difference(y_set)))
    

def ORNOT(x,y):
    print(x,' OR NOT ',y)
    dIds=set(docids)
    
    x_dict=di.get(x)
    y_dict=di.get(y)
    
    x_set=set(x_dict[1])
    y_set=set(y_dict[1])
    universal=set(docids)
    print('\nretrieved docs are:',universal.difference(y_set.difference(x_set)))
    print('\nno.of retrieved docs:',len(universal.difference(y_set.difference(x_set))))
```

```python
def pre_processing_test(term):
    term=lowercase(term)
    term=contract(term)
    term=regTokenize(term)
    term=lemma(term)
    term=stemming(term)
    return term[0]
```
